```python
#estrutura para salvar o mapa de variáveis
variable_map = {}

deriv_df = pd.read_excel("./_ignore_data_censo/Calculo_derivadas_nova - 24072019.xls", header=None)
deriv_df
```

```python
for i, row in deriv_df.iterrows():
    if row[1].strip() in variable_map:
        print(f"Duplicated variable found: {row[1]}")
        continue
    
    variable_data = {}
    variable_data["name"] = row[3]
    variable_data["dtype"] = row[2]
    variable_data["type"] = "condition" if isinstance(row[6], str) and len(row[6].strip()) > 0 else "value"
    variable_data["condition"] = row[6]
    
    variable_map[row[1].strip()] = variable_data
```

Iterando sobre as variáveis calculadas, não encontramos nenhum tipo de duplicação de valores

```python
#Vamos verificar com o arquivo JSON de filtros, se as equações estão corretas
filter_file_path = "data_2017/Metadata/filtros/filtros.json"

filter_data = json.load(open(filter_file_path))['Filtros']
for variable in filter_data:
    var_cod = variable['CodigoVariavel']
    var_condition = variable['Condicao']

    if var_cod not in variable_map:
        print(f"Variable {var_cod} not found in variable map.")
        continue

    if variable_map[var_cod]["condition"] != var_condition:
        print(f"Equation mismatch for variable {var_cod}: {variable_map[var_cod]['condition']} != {var_condition}")
```

```python
def map_fucntion(df, variable_map, reference, add_inf):
    for i, row in df.iterrows():
        if pd.isna(row[0]) or (isinstance(row[0], str) and not row[0].strip()):
            # Trata-se de linhas descrevendos os valores possíveis da variável anterior
            if i == 0:
                print("Skipping first line of dataframe.")
                continue
            
            thresholds = {}
            if last_var in add_inf:
                for var_data in add_inf[last_var]["data"]:
                    if var_data["codigoProduto"] == row[4].strip():
                        thresholds = {
                            "min": var_data.get("limiteInferior", ''),
                            "max": var_data.get("limiteSuperior", ''),
                            "minYield": var_data.get("rendimentoMinimo", ''),
                            "maxYield": var_data.get("rendimentoMaximo", ''),
                            "minDensity": var_data.get("densidadeMinima", ''),
                            "maxDensity": var_data.get("densidadeMaxima", ''),
                            "unit": var_data.get("unidadeDeMedida", ''),
                        }
                        break
            
            variable_map[last_var]["possible_values"].append({
                "cod": row[4] if not pd.isna(row[4]) else 'Vazio',
                "description": row[5],
                "thresholds": thresholds
            })
            
            continue
        
        if row[0].strip() in variable_map:
            last_var = row[0].strip()
            variable_map[last_var].update({
                "possible_values": [],
                "reference": reference
            })
            continue
        
        variable_data = {}
        variable_data["name"] = row[5]
        variable_data["dtype"] = row[4]
        variable_data["type"] = "value"
        variable_data["possible_values"] = []
        variable_data["condition"] = None
        variable_data["reference"] = reference

        variable_map[row[0].strip()] = variable_data
        last_var = row[0].strip()


sheet_names = ["ESTABELEC AGRO", "PECUARIA", "LAVOURA TEMPORARIA", "LAVOURA PERMANENTE",
               "EXTRACAO VEGETAL", "HORTICULTURA", "FLORICULTURA", "EFETIVOS DA SILVICULTURA", 
               "PRODUTOS DA SILVICULTURA", "AGROINDUSTRIA"]
last_rows_skip = [4, 3, 3, 3, 3, 3, 3, 3, 3, 3]
additional_inf = [
    {}, 
    {}, 
    {"V34010100": lavtemp_data}, 
    {"V35010100": lavper_data}, 
    {"V36010100": pextveg_data}, 
    {"V37020100": phort_data}, 
    {"V38010100": pflori_data}, 
    {},
    {"V40010100": ppsil_data}, 
    {"V41010200": agroind_data}
]

for sheet, skip_rows, add_inf in zip(sheet_names, last_rows_skip, additional_inf):
    dict_df = pd.read_excel("data_2017/Dicionario Censo Agropecuário 2017_DERIVADAS.xls", 
                            header=None, sheet_name=sheet, skiprows=7)[:-skip_rows]

    map_fucntion(dict_df, variable_map, sheet, add_inf)
```

```python
questions_file_path = sorted(glob("data_2017/Metadata/quadros/*.json"))

for file_path in questions_file_path:
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    for questions in data.get('Quesitos', []):
        for sub_question in questions["Perguntas"]:
            if sub_question["CodigoVariavel"] in variable_map:
                continue
            
            variable_map[sub_question["CodigoVariavel"]] = {
                "name": sub_question.get("Titulo", "Sem titulo"),
                "dtype": sub_question.get("Tipo", ''),
                "type": sub_question.get("TipoVariavel", 'value')
            }
```
